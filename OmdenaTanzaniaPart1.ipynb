```python
Place = []
Region = []
Country = []
Latitude = []
Longitude = []
WindSpeed = []
WindDegree = []
WindDirection = []
Gust = []
Pressure = []
Precipitation = []
Temperature = []
Visibility = []
Humidity = []
Cloud = []
UV = []

for i in range(df.shape[0]):

    geolocator = Nominatim(user_agent="four_square")
    location = geolocator.geocode(df.City[i] + ', ' + df.Region[i] + ', Tanzania')
    latitude = location.latitude
    longitude = location.longitude

    print(location)
    URL = BASEURL + "/current.json?key=" + APIKEY + "&q=" + df.City[i] + ', ' + df.Region[i] + ', Tanzania' + "&aqi=yes"
    response = requests.get(URL) # HTTP request

    data = response.json()

    Place.append(data["location"]['name'])
    Region.append(data['location']['region'])
    Country.append(data['location']['country'])
    Latitude.append(str(data['location']['lat']))
    Longitude.append(str(data['location']['lon']))
    WindSpeed.append(str(data['current']["wind_mph"]))
    WindDegree.append(str(data['current']["wind_degree"]))
    WindDirection.append(data['current']["wind_dir"])
    Gust.append(str(data['current']["gust_mph"]))
    Pressure.append(str(data['current']["pressure_mb"]))
    Precipitation.append(str(data['current']["precip_mm"]))
    Temperature.append(str(data['current']["feelslike_c"]))
    Visibility.append(str(data['current']["vis_miles"]))
    Humidity.append(str(data['current']["humidity"]))
    Cloud.append(str(data['current']["cloud"]))
    UV.append(str(data['current']["uv"]))

status = pd.DataFrame({'Place': Place,
                           'Region': Region,
                           'Country': Country,
                           'Latitude': Latitude,
                           'Longitude': Longitude,
                           'WindSpeed': WindSpeed,
                           'WindDegree': WindDegree,
                           'WindDirection': WindDirection,
                           'Gust': Gust,
                           'Pressure': Pressure,
                           'Precipitation': Precipitation,
                           'Temperature': Temperature,
                           'Visibility': Visibility,
                           'Humidity': Humidity,
                           'Cloud': Cloud,
                           'UV': UV})

print('Data Shape: ', status.shape)
status.head()
```
